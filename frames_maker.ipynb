```python
import cv2
import os
import csv
```

```python
# Get % of moisture from file name
files = []
videos = []
folder = os.listdir("video_cropped")
for x in folder:
    files.append(x)
for filename in files:
    num = filename.split("_")[0].replace("%", "").replace(".mp4", "")
    moisture = int(num)
    videos.append((filename, moisture))
print(videos)
```

```python
os.makedirs("frames", exist_ok=True)
rows = []  # csv

for video_path, moisture_pct in videos:
    cap = cv2.VideoCapture(f"video_cropped/{video_path}")
    fps = cap.get(cv2.CAP_PROP_FPS)
    frame_idx = 0
    saved = 0

    while True:
        ret, frame = cap.read()
        if not ret:
            break
        if frame_idx % 3 == 0:  # 3 frame per second
            filename = f"{video_path[:-4]}_frame{saved:04d}.jpg"
            cv2.imwrite(f"frames/{filename}", frame)
            rows.append([filename, moisture_pct])
            saved += 1
        frame_idx += 1

    cap.release()
    print(f"{video_path} → {saved} frames saved at {moisture_pct}%")

# Save labels CSV automatically
with open("labels.csv", "w", newline="") as f:
    writer = csv.writer(f)
    writer.writerow(["filename", "moisture"])
    writer.writerows(rows)

print(f"labels.csv created with {len(rows)} total frames")
```
